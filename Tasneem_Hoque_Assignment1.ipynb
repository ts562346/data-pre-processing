```python
import numpy as np
```

```python
import pandas as pd
df = pd.read_csv('data\loan-data.csv')
df.head(10)
```

```python
df.isna().sum()
```

```python
# id
df.sort_values(by="id", inplace=True)
```

```python
# issue_date
df['issue_d'] = pd.DatetimeIndex(df['issue_d']).month
df.rename(columns={'issue_d':'issue_date'}, inplace = True)
```

```python
# read conversion file and make average of all the open, close, high and low column
rate = pd.read_csv("data/EUR-USD.csv")
rate['avg'] = (rate['Open'] + rate['High'] + rate['Low'] + rate['Close'])/4
print(rate)
```

```python
# exchange_rate

df['issue_date'] = df['issue_date'].fillna(1)
df['issue_date'] = df['issue_date'].astype('int')

rate_series = pd.Series(rate.avg.values, index=rate.index.values).to_dict()

def ex_rate(value):
    for month, r in rate_series.items():
        if value == (month+1):
            return r
        
int_grades = df["issue_date"].map(ex_rate)
df["exchange_rate"] = pd.Categorical(
    int_grades
)

df
```

```python
# loan_amnt

df['loan_amnt'] = df['loan_amnt'].astype('float')
df['exchange_rate'] = df['exchange_rate'].astype('float')

df['loan_amnt'] = df['loan_amnt'].fillna(df['loan_amnt'].max()) 

df['loan_amnt_EUR'] = df['loan_amnt'] * df['exchange_rate']
df.rename(columns={'loan_amnt':'loan_amnt_USD'}, inplace = True)
```

```python
# loan_status

df['loan_status'] = df['loan_status'].fillna('Charged Off') 
df['loan_status'] = df['loan_status'].astype('str')

def status(value):
    if value == "Charged Off" or value[0:4] == "Late":
        return 1
    else :
        return 0

loan_status = df["loan_status"].map(status)
df["loan_status"] = pd.Categorical(
    loan_status
)

df.head(20)

# filled empty spaces in this column with 0 indicating 'current or not paid' (assume worse case)
```

```python
# funded_amnt

df['funded_amnt'] = df['funded_amnt'].astype('float')

df['funded_amnt'] = df['funded_amnt'].fillna(df['funded_amnt'].max()) 

df['funded_amnt_EUR'] = df['funded_amnt'] * df['exchange_rate']
df.rename(columns={'funded_amnt':'funded_amnt_USD'}, inplace = True)
```

```python
# term

df['term'] = df['term'].fillna('60 months') # assume worst case here
df['term'] = df['term'].astype('str')

df['term'] = df['term'].str.slice(0,3)
df['term'] = df['term'].astype('int')

df.rename(columns={'term':'term_months'}, inplace = True)
```

```python
# sub_grade

df['sub_grade'] = df['sub_grade'].fillna(df['grade']+'5') # fill sub_grades with worst level of grades
df['sub_grade'] = df['sub_grade'].fillna('36') # fill subgrades with no grades as worst-case

df['sub_grade'] = df['sub_grade'].astype('str')

grades = {
    "A" : 0,
    "B" : 5,
    "C" : 10,
    "D" : 15,
    "E" : 20,
    "F" : 25,
    "G" : 30
}

def grade_mapping(value):
    for key, num in grades.items():
        if value[0] == key:
            return num + int(value[1])
        
int_grades = df["sub_grade"].map(grade_mapping)
df["sub_grade"] = pd.Categorical(
    int_grades
)


# when filling in missing grade values, replace empty spaces with 36
```

```python
group = df[["int_rate", "sub_grade"]].groupby("sub_grade").max()
group
```

```python
# int_rate

df['int_rate'] = df['int_rate'].fillna(df["sub_grade"]) 
df ['int_rate'] = df ['int_rate'] / 100
```

```python
# installment

df['installment'] = df['installment'].astype('float')

df['installment'] = df['installment'].fillna(df['loan_amnt']*df['int_rate']/df['term']) 

df['installment_EUR'] = df['installment'] * df['exchange_rate']
df.rename(columns={'installment':'installment_USD'}, inplace = True)
```

```python
# grade
df.drop(columns=["grade"], inplace=True)
```

```python
# verification_status

df['verification_status'] = df['verification_status'].astype('str')

def isVerified(value):
    if value == "Not Verified" or value == "nan":
        return 1
    else :
        return 0

ver_status = df["verification_status"].map(isVerified)
df["verification_status"] = pd.Categorical(
    ver_status
)

# filled empty spaces in this column with 0 indicating 'not verified' (assume worse case)
```

```python
# url
df.drop(columns=["url"], inplace=True)
```

```python
# addr_state

regions = {
    "East" : ["PA", "NY", "VT", "NH", "ME", "MA", "RI", "CT", "NJ"],
    "West" : ["WA", "OR", "CA", "NV", "ID", "MT", "WY", "UT", "AZ", "NM", "CO", "AK"],
    "Midwest" : ["ND", "SD", "NE", "KS", "MN", "IA", "MO", "IL", "WI", "IN", "MI", "OH"],
    "South" : ["OK", "TX", "AR", "LA", "MS", "TN", "KY", "AL", "FL", "GA", "SC", "NC", "VA", "WV", "DC", "MD",
              "DE", "HI"]
}

df['addr_state'] = df['addr_state'].fillna('CA') 

def region_mapping(value):
    for region, states in regions.items():
        for state in states:
            if value == state:
                return region

reg_map = df["addr_state"].map(region_mapping)
df["addr_state"] = pd.Categorical(
    reg_map
)

df.rename(columns={'addr_state':'state_address'})
```

```python
# total_pymnt

df['total_pymnt'] = df['total_pymnt'].astype('float')

df['total_pymnt'] = df['total_pymnt'].fillna(0) 

df['total_pymnt_EUR'] = df['total_pymnt'] * df['exchange_rate']
df.rename(columns={'total_pymnt':'total_pymnt_USD'}, inplace = True)
```
